```python
import matplotlib.pyplot as plt

x = [0, 2, 4, 6, 8]
y = [0, 4, 16, 36, 64]

fig, ax = plt.subplots()  
ax.plot(x, y, marker='o', label="Data Points")

ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-Axis") 
ax.set_ylabel("Y-Axis")  


plt.show()
```

```python
import numpy as np

# 1. Verificar a Versão do NumPy
print(f"Versão do NumPy instalada: {np.__version__}\n")

# 2. Criar duas matrizes (arrays NumPy) 2x2
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# 3. Realizar uma operação básica (Multiplicação de Matrizes - Produto Escalar)
C = np.dot(A, B)

# 4. Exibir os resultados
print("Matriz A:")
print(A)

print("\nMatriz B:")
print(B)

print("\nProduto Escalar (A @ B):")
print(C)
```

```python
from sympy import symbols, diff, exp

# 1. Definir a variável como um símbolo
x = symbols('x')

# 2. Definir uma expressão matemática (Função)
# f(x) = x^2 * e^x
f_x = x**2 * exp(x)

# 3. Realizar o Cálculo Simbólico (Derivada)
# Calcula a derivada de f_x em relação a x
derivada = diff(f_x, x)

# 4. Exibir os resultados
print(f"Variável Simbólica: {x}")
print(f"Expressão Original (f(x)): {f_x}")
print(f"Derivada (f'(x)): {derivada}")
```

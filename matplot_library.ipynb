```python
#Bar chart

import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60]
y = [13, 45, 23, 34, 96, 76]
plt.title('Bar Graph')
plt.bar(x, y, color='dodgerblue', width=5)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt # Importar matplotlib para exibir o grfico

# 1. Definio dos vetores de dados (eixos X e Y)
progressao_km = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900]
altimetria_via = [85, 92, 98, 105, 101, 95, 88, 80, 75, 70]

# 2. Criao de um DataFrame Pandas a partir dos vetores
# O Seaborn precisa dos dados neste formato
dados_projeto = pd.DataFrame({
    'Progressao_KM': progressao_km,
    'Altimetria_M': altimetria_via
})

# 3. Gerar o grfico de linhas usando o Seaborn
plt.figure(figsize=(10, 6)) # Define o tamanho da figura para melhor visualizao

# Usando o sns.lineplot com os nomes das colunas do nosso DataFrame
sns.lineplot(x='Progressao_KM', y='Altimetria_M', data=dados_projeto)

# 4. Adicionar ttulos e rtulos para contexto de Engenharia Civil
plt.title('Perfil Longitudinal da Via (Altimetria vs. Progresso)')
plt.xlabel('Progresso Quilomtrica (Metros)')
plt.ylabel('Cota (Elevao em Metros)')
plt.grid(True) # Adiciona uma grade para facilitar a leitura

# 5. Mostrar o grfico
plt.show()
```
